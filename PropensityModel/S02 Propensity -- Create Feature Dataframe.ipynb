## Preprocessing

```
from pyspark.sql import functions as f
from pyspark.sql import types as t
from pyspark.sql import SparkSession as SparkSession
# from featureGen.numericalProcessing.createFeatureDataframes.createBehaviorFeatures
from featureGen.numericalProcessing.createFeatureDataframes.createCategoryDummies import createCategoryDummies as createCategoryDummies
from geoCode.geoHash import geoHashSpark as geoHashSpark
from pyspark.sql import Window
```

```
countryCode = 'de'
```

```
universe = spark.table("dev_derived_ouniverse." + "output_total_universe_" + countryCode) \
                                                .withColumn('longitude',f.col('longitude').cast('float')) \
                                                .withColumn('latitude',f.col('latitude').cast('float')) \
                                                .where((f.col('latitude').isNotNull()) &(f.col('longitude').isNotNull())) \
                                                .withColumn('dep_dum_buyer',f.when(f.col('ohubid').isNotNull(),f.lit(1)).otherwise(f.lit(0))) \

universe = geoHashSpark(universe,'latitude','longitude',5)
2wqasz
features  = spark.table("data_user_hien." + countryCode + "_geo_data") \
                            .withColumn('popPerHash', f.when(f.col('popPerHash').isNotNull(),f.col('popPerHash')).otherwise(f.lit(0))) \
                            .withColumn('buildPerHash', f.when(f.col('buildPerHash').isNotNull(),f.col('buildPerHash')).otherwise(f.lit(0))) \

```

```
display(universe)
```

```
countBusinesses             = universe.groupBy('geohash').agg(f.count(f.col('operatorID')).alias('businessesPerHash'))

universeWithCompetition     = universe.alias('u').join(countBusinesses.alias('b'),universe.geoHash == countBusinesses.geohash, how = 'left_outer').select('u.*','b.businessesPerHash')

universeWithCompetitionPopUrban = universeWithCompetition.alias('de').join(features.alias('f'),universeWithCompetition.geoHash == features.geoHash, how = 'left_outer').select('de.*','f.popPerHash','f.buildPerHash')
```

```
universeWithCompetitionPopUrban = universeWithCompetitionPopUrban\
                                    .withColumn("popPerHash",f.when(f.col('popPerHash').isNotNull(),f.col('popPerHash')).otherwise(f.lit(0)))\
                                    .withColumn('buildPerHash', f.when(f.col('buildPerHash').isNotNull(),f.col('buildPerHash')).otherwise(f.lit(0))) 
```

```
from pyspark.ml.feature import QuantileDiscretizer
from pyspark.ml.feature import Bucketizer

def decile(inCol,outCol,df):
  
  discretizer = QuantileDiscretizer(numBuckets=10, inputCol=inCol,outputCol=outCol)  
  dfOut = discretizer.fit(df).transform(df)
  dfOut = dfOut.withColumn(outCol,f.col(outCol) + 1 )
  return dfOut

universeWithCompetitionPopUrban = decile('buildPerHash','urbanisationLevel',universeWithCompetitionPopUrban)
universeWithCompetitionPopUrban = decile('popPerHash','densityLevel',universeWithCompetitionPopUrban)
universeWithCompetitionPopUrban = decile('businessesPerHash','competitionLevel',universeWithCompetitionPopUrban)
```

## Create Dummmies Feature Dataframe

```
universeWithCompetitionPopUrban = universeWithCompetitionPopUrban.withColumn('urbanisationLevel',f.col('urbanisationLevel').cast('int'))\
                                                                 .withColumn('densityLevel',f.col('densityLevel').cast('int'))\
                                                                 .withColumn('competitionLevel',f.col('competitionLevel').cast('int'))\
                                                                 .withColumnRenamed('popPerHash','ind_popPerHash') \
                                                                 .withColumnRenamed('businessesPerHash','ind_businessesPerHash')\
                                                                 .withColumnRenamed('buildPerHash','ind_buildPerHash')\
```

```
spaceDeleteUDF = f.udf(lambda s: s.replace(" ", "_"), t.StringType())
universeWithCompetitionPopUrban = universeWithCompetitionPopUrban.withColumn("globalChannel", spaceDeleteUDF("globalChannel"))
universeWithCompetitionPopUrban = universeWithCompetitionPopUrban.withColumn("globalChannel", f.regexp_replace(f.col("globalChannel"), "[^\w\s]",""))
```

```
universeWithCompetitionPopUrban.write.mode('overwrite').saveAsTable('data_user_hien.psm_universe_' + countryCode)
```

```
dfTempTypeDummies =  createCategoryDummies(dataframe = universeWithCompetitionPopUrban,
                            feature = "globalChannel",
                            ID ="operatorid",
                            lower_bound = 30,
                            prefix = 'ind_dum_business_type_') 
```

```
dfTempUrbanisationDummies = createCategoryDummies(dataframe = universeWithCompetitionPopUrban,            
                            feature = "urbanisationLevel",                       
                            ID = "operatorid",             
                            lower_bound = 30,                                    
                            prefix = "ind_dum_geo_urbanisation_")
```

```
dfTempCompetitionDummies = createCategoryDummies(dataframe = universeWithCompetitionPopUrban,            
                            feature = "competitionLevel",                       
                            ID = "operatorid",             
                            lower_bound = 30,                                    
                            prefix = "ind_dum_competition_")
```

```
dfTempDensityDummies =  createCategoryDummies(dataframe = universeWithCompetitionPopUrban,            
                            feature = "densityLevel",                       
                            ID = "operatorid",             
                            lower_bound = 30,                                    
                            prefix = "ind_dum_density_")
```

```
dfFeatures   = dfTempTypeDummies.join(dfTempUrbanisationDummies, on = ['operatorId'], how = 'left')\
                                .join(dfTempCompetitionDummies, on = ['operatorId'], how = 'left')\
                                .join(dfTempDensityDummies, on = ['operatorId'], how = 'left')\

dfFeatures.select('operatorId').distinct().count()
```

```
from pyspark.sql import functions as f
from pyspark.sql import types as t
import dateutil.parser 
```

```
countryCode = 'za'
```

#### Read relevant tables

```
universe       =  spark.table('dev_derived_ouniverse.output_total_universe_' +countryCode)\
                       .withColumn('d_buyer',f.when(f.col("operatorOhubId").isNotNull(),f.lit(1)).otherwise(f.lit(0)))

sale           =  spark.table("dev_sources_ohub.cleaned_operator_sales") \
                       .where(f.lower(f.col('countryCode')) == countryCode) \
                       .where(f.col('transactionDate').isNotNull()) \
                       .where(f.col('cuEanCode').isNotNull()) \
                       .where(f.col('amount') > 0) \
                       .withColumn('year',f.substring(f.col('transactiondate'),1,4).cast('int')) \
                       .where(f.col('year') >= 2019)


```

```
display(sale)
```

#### Create table with relevant products per operator based on the sales data

##### Get Relevant products for operator from the sales table

```
universeRelevantProductsSSD = universe.alias('u').join(sale.alias('s'), on = ['operatorOhubID'], how ='inner').\
                                      select('u.operatorid','s.cuEanCode','s.productName', 's.quantityOfUnits', 's.cuListingPriceInCents').\
                                      where((f.col('quantityOfUnits').isNotNull()) & (f.col('cuListingPriceInCents').isNotNull())).\
                                      withColumn('totalProductPrice', f.col('quantityOfUnits')*(f.col('cuListingPriceInCents'))).\
                                      withColumn('totalProductPrice', f.col('totalProductPrice')/f.lit(100)).\
                                      distinct()
```

##### Create Assumptions table

```
productAssumptions = universeRelevantProductsSSD.groupBy('cuEanCode').agg(f.avg(f.col('quantityOfUnits')).alias('avgQuantityOfUnits'),\
                                                                          f.avg(f.col('cuListingPriceInCents')).alias('avgListingPriceInCents'),\
                                                                          f.avg(f.col('totalProductPrice')).alias('avgTotalProductPrice'))
```

```
display(productAssumptions)
```

```
print(
      'relevant products assumptions table row count: ', productAssumptions.count(), '\n',
      'relevant products assumptions table id count: ', productAssumptions.select('cuEanCode').distinct().count()
     )
```

```
productAssumptions.write.mode('overwrite').saveAsTable('data_user_hien.valueTier_productAssumptions_' + countryCode)
```
